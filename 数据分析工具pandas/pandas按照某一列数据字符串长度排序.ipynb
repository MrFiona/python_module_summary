# pandas按照某一列数据字符串长度排序

```python
import pandas as pd
df = pd.read_csv('pandas_test.csv')
```

```python
df
```

```python
#todo 第一列Team数据长度Series
Team_length_series = df['Team'].apply(lambda x: len(x))
```

```python
#todo 在Team列之后增加长度列以便后续排序
df.insert(1, 'Team_length', Team_length_series)
```

```python
#todo 按照Team_length列排序
df.sort_values(by='Team_length', inplace=True)
```

```python
df
```

```python
df[df['Team'].str.len() < 7]
```

```python
df[df['Team'].map(len) < 7]
```

```python
x1 = -1
m = df['Team'].values
m = sorted(m, key=len, reverse=False)
print(m, type(m))
def func(x):
    global x1
    x1 += 1
    return m[x1]

df['Team'].apply(func)
```
